```python
def calcularIva(monto_prestamo):
    return monto_prestamo * 0.19
   

df["IVA"]= df["MONTO_DEL_PRESTAMO"].apply(calcularIva)# Se aplica la funcion a la fila
df.head()
```

```python
#df.pop("IVA") --> Borra y lista el contenido
def calcularInteresAlDosP(fila):
    return fila["IVA"]+fila["MONTO_DEL_PRESTAMO"]
df["INTERES_AL_DOS_PORCIENTO"]= df.apply(calcularInteresAlDosP, axis=1)
#axis es que tiene que ejecutar la operacion por todas las filas
df.head()
    
```

```python
#AGRUPAR DATOS
df.groupby("MONTO_DEL_PRESTAMO").mean()
```

```python
df.groupby("MONTO_DEL_PRESTAMO").mean({
    'IVA':'sum',
    'INTERES_AL_DOS_PORCIENTO':'mean'
    #'campo':'operacion ejemplo: max o min'
})
```

```python
grouped = df.groupby("MONTO_DEL_PRESTAMO").agg({
    'IVA':'sum',
    'INTERES_AL_DOS_PORCIENTO':'mean'
    #'campo':'operacion ejemplo: max o min'
})
grouped[grouped["IVA"]>55404.0]
```

```python
import matplotlib.pyplot as plt

grouped["INTERES_AL_DOS_PORCIENTO"].head(10).plot(kind="barh")
plt.show()
```

```python
#Exportar grouped to csv
grouped.to_csv("interes_alDosPorciento.csv")
```
