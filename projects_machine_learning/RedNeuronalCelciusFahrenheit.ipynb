# REDES NEURONALES
Una red neuronal se compone de neuronas artificiales que se comportan de
forma similar a una ***neurona cerebral***.
Al igual que las naturales, las ***neuronas artificiales*** contienen ***ramificaciones*** y
***un núcleo o nodo***, que corresponden a las entradas de la neurona que
provienen de otras neuronas. La comunicación entre neuronas se realiza a
través de las ramificaciones.


RedNeuronalExample.png

- Las redes neuronales se componen de capas, y cada capa puede tener una
  o más neuronas.
- Toda red neuronal tiene al menos una capa de entrada y una de salida.
- Las capas intermedias se conocen como capas ocultas.
- Las neuronas se comunican a través de conexiones y cada conexión tiene
  un peso asignado. El peso es un valor numérico que hace referencia a la
  importancia de la conexión entre las neuronas.
- Cada neurona, excepto las de entrada, tienen un sesgo, que también es un valor numérico.

## Import the libraries

````python
import tensorflow  as tf
import numpy as np
````

## Provided Data Already Calculated

````python
celsius = np.array([-40,-20,-10,0,10,15,20,25,30,35], dtype=float)
fahrenheit = np.array([-40,-4,14,32,50,59,68,77,86,95], dtype=float)
````

## Create The Network Neuronal Model

- Capas densas son las que tiene conexiones de **todas** las neuronas
- units : unidades o neuronas de la capa de salida
- input_shape: unidades o neuronas de la capa de entrada




````python
capa = tf.keras.layers.Dense(units=1, input_shape = [1])
modelo  = tf.keras.Sequential([capa])
````

## Compile the Model

- El modelo **Adam** permite ajustar pesos y cargas de manera eficaz
- De manera progresiva vaya mejorando
- El parámetro del modelo Adam es la tasa de aprendizaje 
- loss = función de pérdida


````python
modelo.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
````

## Train the model

- epochs  = ciclos que realizará la red neuronal ( Prueba y Error)
- verbose = False: no imprime tanta información

````python
Data_Train = modelo.fit(celsius, fahrenheit, epochs=500, verbose=False)
print('Modelo Exitosamente Entrendado !!! ')
````

## Plot the loss function

````python
import matplotlib.pyplot as plt
plt.xlabel("Iteration")
plt.ylabel("Magnitude of Loss")
plt.plot(Data_Train.history["loss"])
plt.show()
````

## Predictions

````python
print("-----------------Predictions-------------------")
y_pred = modelo.predict([100])
print("El resultado es", str(y_pred), "°F")
````

## Print the variables

````python
print("Peso y sesgo: ", capa.get_weights())
````

## Add more layers

````python
oculta1 = tf.keras.layers.Dense(units = 5, input_shape= [1])
oculta2 = tf.keras.layers.Dense(units = 5)
salida = tf.keras.layers.Dense(units = 2)
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
````

````python
modelo.compile(optimizer= tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
````

capa2.png

## Train the model second time

````python
Data_Train = modelo.fit(celsius, fahrenheit, epochs = 600, verbose = False)
print("Modelo Entrenado !! x2")
````

## Plot loss function second time

````python
import matplotlib.pyplot as plt
plt.xlabel("Iteration")
plt.ylabel("Magnitude of Loss")
plt.plot(Data_Train.history["loss"])
plt.show()
````

## Faster Learning

````python
print("--------------PREDICTIONS--------------")
y_pred2= modelo.predict([100])
print("El resultado es: ", str(y_pred2), "°F")
````
